```python
# variables globales 
mon_score = 0
score_pc = 0

import tkinter as tk
from random import randint  
    
window = tk.Tk()
window.geometry("500x400")
window.config(bg="#FFDDDD")
window.title("Jeu Pierre Papier Ciseaux")
window.resizable(width=False, height=False)

def Exit():
    window.destroy()
    
def play(mon_choix,comp_pick): 
    
    mon_choix=ent_choix.get()
        
    global mon_score, score_pc    
    if(mon_choix == comp_pick):
        lbl_resultat.configure(text="Egalité!") 
    elif(mon_choix == "pierre") and (comp_pick == "papier"):
        lbl_resultat.configure(text="Perdu!") 
        score_pc = score_pc + 1
    elif(mon_choix == "papier") and (comp_pick == "pierre"):
        lbl_resultat.configure(text="Gagné!") 
        mon_score = mon_score + 1
    elif(mon_choix == "papier") and (comp_pick == "ciseaux"):
        lbl_resultat.configure(text="Perdu!") 
        score_pc = score_pc + 1
    elif (mon_choix == "ciseaux") and (comp_pick == "papier"):
        lbl_resultat.configure(text="Gagné!") 
        mon_score = mon_score + 1
    elif(mon_choix == "ciseaux") and (comp_pick == "pierre"):
        lbl_resultat.configure(text="Perdu!") 
        score_pc = score_pc + 1
    elif (mon_choix == "pierre") and (comp_pick == "ciseaux"):
        lbl_resultat.configure(text="Gagné!") 
        mon_score = mon_score + 1
    else:
        lbl_resultat.configure(text="Vérifiez l'orthographe de votre choix!") 
jeu  = ["pierre", "papier", "ciseaux"]

def au_hasard():
    comp_pick=jeu[randint(0,2)]
    lbl_ord.configure(text=comp_pick)
    play(mon_choix, comp_pick)
    lbl_mon_score.configure(text=str(mon_score))    
    lbl_score_pc.configure(text=str(score_pc)) 

def Reset():    
    global mon_score, score_pc    
    score_pc = 0    
    mon_score = 0
    lbl_mon_score.configure(text=str(mon_score))    
    lbl_score_pc.configure(text=str(score_pc)) 
    ent_choix.delete(0, 'end')
    lbl_ord.configure(text="")
    lbl_resultat.configure(text="") 
    
        
l1_title = tk.Label(window,text="Jeu Pierre Papier Ciseaux",font=("Arial", 15),fg="white",bg="#A569BD" )
l1_title.place(x=120, y=20)

lbl_empty=tk.Label(window, bg="#FFDDDD")
lbl_empty.grid(row=0, column=0, pady=30)


lbl_moi = tk.Label(window,text = "Moi" , font={13}, bg="#FFDDDD")
lbl_moi.grid(row=1, column=0)
lbl_pc = tk.Label(window, text= "Ordinateur" , font={13}, bg="#FFDDDD")
lbl_pc.grid(row=1, column=1)

lbl_mon_score = tk.Label(text="0", font={13}, bg="#FFDDDD")
lbl_mon_score.grid(row=2, column=0)
lbl_score_pc = tk.Label(text="0", font={13}, bg="#FFDDDD")
lbl_score_pc.grid(row=2, column=1)

lbl_empty2=tk.Label(window,bg="#FFDDDD")
lbl_empty2.grid(row=3, column=0, pady=15)

lbl_choix = tk.Label(text="Choisir entre pierre papier ciseaux", font={13}, bg="#FFDDDD")
lbl_choix.grid(row=4, column=0)
ent_choix = tk.Entry(width=15, font={13})
ent_choix.grid(row=4, column=1)


btn_ordinateur = tk.Button(master=window, text="choix de l'ordinateur  ", font={13}, command=au_hasard)
btn_ordinateur.grid(row = 5, column=0, pady=15)
lbl_ord = tk.Label(master=window, font={13}, bg="#FFDDDD", text="")
lbl_ord.grid(row = 5, column=1)

lbl_resultat = tk.Label(master=window, font={13}, bg="#FFDDDD", text="", fg="red")
lbl_resultat.place(x= 220, y=300)

btn_rejouer = tk.Button(master=window, text="Rejouer", font={13}, command=Reset)
btn_rejouer.place(x=160, y=350)

btn_rejouer = tk.Button(master=window, text="Quitter", font={13}, command=Exit)
btn_rejouer.place(x=280, y=350)

window.mainloop()
```
